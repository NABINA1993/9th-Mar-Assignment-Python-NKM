Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with 
an example.

Probability Mass Function (PMF) and Probability Density Function (PDF) are both mathematical tools used in probability theory and statistics to describe the probability distribution of a random variable.

A Probability Mass Function (PMF) is used to describe the probability distribution of a discrete random variable. It assigns a probability to each possible value that the random variable can take. For example, consider a fair six-sided die. The PMF for this die would assign a probability of 1/6 to each of the possible outcomes (1, 2, 3, 4, 5, or 6) and a probability of 0 to any value outside this range.

On the other hand, a Probability Density Function (PDF) is used to describe the probability distribution of a continuous random variable. It gives the probability of the variable taking on any given value within a range of possible values. The PDF represents the area under the curve of the function and the total area is equal to 1. For example, consider the height of adult men in a population. The PDF for height would represent the probability of a randomly selected man having a height within a certain range, say between 5'6" and 6'0".

Both PMF and PDF provide a way to calculate probabilities associated with different values of a random variable. The main difference between them is that PMF is used for discrete random variables while PDF is used for continuous random variables.





Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Cumulative Density Function (CDF) is a function that gives the cumulative probability that a random variable takes a value less than or equal to a given value. It is defined for both discrete and continuous random variables.

For a discrete random variable, the CDF is defined as the sum of the probabilities of all outcomes less than or equal to the given value. For a continuous random variable, the CDF is defined as the integral of the probability density function (PDF) from negative infinity to the given value.

CDF is used to determine the probability of a random variable taking a value less than or equal to a given value. It can also be used to determine the probability of a random variable taking a value within a particular range.

Example: Let's consider a fair six-sided die roll. The possible outcomes are 1, 2, 3, 4, 5, and 6, each with a probability of 1/6. The CDF for this random variable is given by:

For any x between 1 and 6 (inclusive), the CDF of the die roll is:

CDF(x) = P(X <= x)

If we want to find the probability of getting a number less than or equal to 3, we can calculate the CDF as follows:

CDF(3) = P(X <= 3)
= P(X = 1) + P(X = 2) + P(X = 3)
= 1/6 + 1/6 + 1/6
= 1/2

Therefore, the probability of getting a number less than or equal to 3 is 1/2.





```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random dataset
data = np.random.normal(0, 1, size=1000)

# Calculate the CDF using numpy
counts, bin_edges = np.histogram(data, bins=100, density=True)
cdf = np.cumsum(counts) * np.diff(bin_edges)

# Plot the CDF using matplotlib
plt.plot(bin_edges[1:], cdf)
plt.xlabel('Data Values')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Density Function (CDF)')
plt.show()
```

Explain how the parameters of the normal distribution relate to the shape of the distribution.

The normal distribution is commonly used as a model in many real-world situations. Some examples include:

Heights and weights of a population: The normal distribution can be used to model the distribution of heights and weights in a population, with the mean and standard deviation representing the average height and variability, respectively.

Test scores: The normal distribution can be used to model the distribution of test scores, with the mean and standard deviation representing the average score and variability, respectively.

Measurement errors: The normal distribution can be used to model measurement errors in scientific experiments, with the mean and standard deviation representing the true value of the measured quantity and the precision of the measurement, respectively.

The parameters of the normal distribution are the mean (μ) and the standard deviation (σ). The mean represents the central tendency of the distribution, while the standard deviation represents the spread or variability of the distribution. A higher standard deviation indicates that the data is more spread out, while a lower standard deviation indicates that the data is more clustered around the mean. The shape of the distribution is symmetric around the mean, with the peak of the distribution occurring at the mean. The spread of the distribution is determined by the standard deviation - a larger standard deviation results in a wider distribution, while a smaller standard deviation results in a narrower distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the normal distribution
mu = 0  # mean
sigma = 1  # standard deviation

# Generate 1000 random numbers from the normal distribution
samples = np.random.normal(mu, sigma, 1000)

# Plot the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.5)

# Plot the probability density function (PDF) of the normal distribution
x = np.linspace(-5, 5, 100)
pdf = 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2/(2*sigma**2))
plt.plot(x, pdf, 'r-', lw=2)

# Plot the cumulative distribution function (CDF) of the normal distribution
cdf = (1 +erf((x - mu) / (sigma * np.sqrt(2)))) / 2
plt.plot(x, cdf, 'g--', lw=2)

plt.legend(['PDF', 'CDF', 'Samples'])
plt.show()
```

Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli 
Distribution and Binomial Distribution?

Bernoulli distribution is a discrete probability distribution that models the outcomes of a single experiment or trial that can result in two possible outcomes, usually referred to as success and failure. The probability of success is denoted by the parameter p and the probability of failure is q=1-p. The Bernoulli distribution has a probability mass function given by:

P(X=x) = p^x (1-p)^(1-x)

where x is the outcome of the experiment (0 for failure and 1 for success).

An example of Bernoulli distribution can be the result of a coin toss experiment where the success can be defined as getting heads and failure as getting tails. Here, the parameter p would be 0.5 as the coin is fair.

The main difference between Bernoulli and Binomial distribution is that Bernoulli distribution models the outcome of a single experiment while Binomial distribution models the number of successes in a fixed number of independent Bernoulli trials. In other words, Binomial distribution is the sum of multiple independent Bernoulli distributions.

Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset 
is normally distributed, what is the probability that a randomly selected observation will be greater 
than 60? Use the appropriate formula and show your calculations.

To solve this problem, we can use the standard normal distribution table and z-score formula.

The z-score formula is given by:

z = (x - μ) / σ

where x is the observation (in this case, 60), μ is the mean (50), and σ is the standard deviation (10).

Substituting the values, we get:

z = (60 - 50) / 10
z = 1

Using the standard normal distribution table, we can find the probability that a z-score is less than 1. This is equivalent to finding the probability that a randomly selected observation will be less than 60.

The table gives us a probability of 0.8413 for a z-score of 1.

Since the normal distribution is symmetric, we know that the probability of a z-score being greater than 1 is equal to the probability of a z-score being less than -1.

Therefore, the probability of a randomly selected observation being greater than 60 is:

P(x > 60) = 1 - P(x < 60)
P(x > 60) = 1 - 0.8413
P(x > 60) = 0.1587

Therefore, the probability that a randomly selected observation will be greater than 60 is 0.1587 or approximately 16%.

```python
import scipy.stats as stats

# define the mean and standard deviation
mean = 50
sd = 10

# calculate the z-score for x = 60
z = (60 - mean) / sd

# calculate the probability using the CDF
prob = 1 - stats.norm.cdf(z)

print("The probability that a randomly selected observation will be greater than 60 is:", prob)
```

Q7: Explain uniform Distribution with an example.

Uniform distribution is a continuous probability distribution where every possible value within a given interval has an equal likelihood of occurring. This distribution is often used to model situations where there is no preference for one outcome over another.

For example, suppose you have a six-sided die, and you roll it once. The probability of getting any particular number (1, 2, 3, 4, 5, or 6) is the same, and each outcome is equally likely. This situation can be modeled using a uniform distribution.

Mathematically, the probability density function of a uniform distribution is defined as:

f(x) = 1 / (b-a) for a ≤ x ≤ b
where a is the lower limit of the interval, b is the upper limit, and f(x) is the probability density function.

For instance, suppose we want to simulate the roll of a fair six-sided die using a uniform distribution. We can define the lower limit a as 1 and the upper limit b as 6. The probability density function of the uniform distribution for this example would be:

f(x) = 1/6 for 1 ≤ x ≤ 6

In this example, the probability of rolling any particular number is 1/6, and all outcomes are equally likely.

Q8: What is the z score? State the importance of the z score.

The z score (also known as the standard score) is a measure of how many standard deviations an observation or data point is away from the mean of the population or sample. It is calculated as the difference between the observed value and the population or sample mean, divided by the standard deviation of the population or sample.

The importance of the z score lies in its ability to standardize data and allow comparisons between different datasets with different means and standard deviations. A z score of 0 indicates that the observation is equal to the mean, while a z score of +1 (or -1) indicates that the observation is one standard deviation above (or below) the mean. A z score of +2 (or -2) indicates that the observation is two standard deviations above (or below) the mean, and so on.

The z score is commonly used in statistical analysis, particularly in hypothesis testing and confidence intervals. It helps to determine the probability of obtaining an observation or sample mean that is within a certain range of values, and to make comparisons between different samples or populations.

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

The Central Limit Theorem (CLT) is a fundamental concept in statistics that describes the behavior of the mean of a random sample drawn from any population. It states that as the sample size increases, the sampling distribution of the sample means approaches a normal distribution, regardless of the shape of the population distribution.

The significance of the Central Limit Theorem is that it provides a mathematical foundation for many statistical techniques that rely on the normal distribution, such as hypothesis testing and confidence intervals. It also allows us to use the normal distribution as an approximation for the distribution of sample means, which simplifies calculations and makes statistical inference more efficient.

Q10: State the assumptions of the Central Limit Theorem.

The assumptions of the Central Limit Theorem (CLT) are:

Independence: The random variables in the sample should be independent of each other.

Sample Size: The sample size should be sufficiently large. There is no hard and fast rule for what "sufficiently large" means, but a general guideline is that the sample size should be at least 30.

Finite Variance: The random variables in the population should have a finite variance. If the variance is infinite, then the CLT does not hold.

Identical Distribution: The random variables in the sample should be identically distributed. This means that they should have the same mean and variance.

If these assumptions are met, then the CLT states that the sampling distribution of the sample mean will approach a normal distribution as the sample size increases, regardless of the distribution of the original population.
